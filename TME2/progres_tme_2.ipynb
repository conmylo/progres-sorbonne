3. Use regular expressions to remove problematic zeros from an IPv4 address expressed as a 
string. (example: "216.08.094.196" should become "216.8.94.196", but "216.80.140.196" 
should remain "216.80.140.196"). 

```python
# Define the function
def normalize_ip(txt):
    # Regular expression pattern to match and remove leading zeros
    pattern = r'\b0+(\d+)\b'
    
    # Store result in variable
    result = re.sub(pattern, r'\1', txt)
    
    return result
```

```python
# Call function for IP not normalized
normalize_ip("216.08.094.196")
```

```python
# Call the function for IP normalized
normalize_ip("216.80.140.196")
```

4. Use regular expressions to transform a date from MM-DD-YYYY format to DD-MM-YYYY 
format. (example "11-06-2020" should become "06-11-2020"). Optionally*, do the same thing using the `datetime` package.

```python
# Define the function
def switch_md(txt):
    # Regular expression pattern for MM-DD-YYYY
    pattern = r'(\d{2})-(\d{2})-(\d{4})'
    
    # Store the result into a variable
    result = re.sub(pattern, r'\2-\1-\3', txt)
    
    return result
```

# Exercice 2 - Analyze XML

Write a Python program that will: 

1. Retrieve the content of the page at:

```python
url = "https://www.w3schools.com/xml/cd_catalog.xml"
```

Look at the text content and load as xml.

```python
import requests
import xml.etree.ElementTree as ET

url = "https://www.w3schools.com/xml/cd_catalog.xml"
output_file = "cd_catalog.xml"

try:
    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Retrieve the content of the page
        page_content = response.text

        # Save the content to an XML file
        with open(output_file, "w", encoding="utf-8") as xml_file:
            xml_file.write(page_content)

        print(f"XML content saved to {output_file}")

    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

# Handle exceptions
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")

except ET.ParseError as e:
    print(f"Error parsing the XML: {e}")
```

2. Display for each CD: title, artist, country, company, year. 

```python
import xml.etree.ElementTree as ET

# Define the function
def display_cd(cd_element):
    try:
        title = cd_element.find("TITLE").text
        artist = cd_element.find("ARTIST").text
        country = cd_element.find("COUNTRY").text
        company = cd_element.find("COMPANY").text
        year = cd_element.find("YEAR").text

        print("CD Information:")
        print(f"Title: {title}")
        print(f"Artist: {artist}")
        print(f"Country: {country}")
        print(f"Company: {company}")
        print(f"Year: {year}")
        print()

    # Handle exceptions
    except AttributeError as e:
        print(f"Error displaying CD information: {e}")

# Usage:
xml_file_path = "cd_catalog.xml"
tree = ET.parse(xml_file_path)
root = tree.getroot()
for cd in root.findall("CD"):
    display_cd(cd)
```

```python
import requests
import json

url = "https://opendata.paris.fr/explore/dataset/lieux-de-tournage-a-paris/download/?format=json&timezone=Europe/Berlin&lang=fr"
output_file = "paris_filming_locations.json"

try:
    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the content as JSON
        json_data = response.json()

        # Save the JSON data to a file
        with open(output_file, "w") as file:
            json.dump(json_data, file, indent=4)

        # Count the number of entries in the JSON data
        entry_count = len(json_data)

        print(f"JSON data downloaded and saved as {output_file}")
        print(f"Number of entries: {entry_count}")

    else:
        print(f"Failed to retrieve the file. Status code: {response.status_code}")

# Handle exceptions
except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")

except json.JSONDecodeError as e:
    print(f"Error parsing JSON data: {e}")
```

```python
import json

json_file_path = "paris_filming_locations.json"  # Replace with the path to your JSON file

try:
    # Load the JSON data from the file
    with open(json_file_path, "r", encoding="utf-8") as file:
        json_data = json.load(file)

    # Create a dictionary to store the information
    movie_data = {}

    # Group entries by movie
    for entry in json_data:
        fields = entry.get("fields")
        if fields:
            director = fields.get("nom_realisateur")
            title = fields.get("nom_tournage")
            shooting_dates = (fields.get("date_debut"), fields.get("date_fin"))
            location = fields.get("ardt_lieu")

            # Create or update movie data
            if title not in movie_data:
                movie_data[title] = {
                    "director": director,
                    "shooting_dates": [shooting_dates],
                    "locations": [location],
                }
            else:
                movie_data[title]["shooting_dates"].append(shooting_dates)
                movie_data[title]["locations"].append(location)

    # Display information for all movies
    counter = 0 # counter to display just 20 results
    for movie, data in movie_data.items():
        if counter >= 20:
            break
        print(f"Movie: {movie}")
        print(f"Director: {data['director']}")
        print("Shooting Dates:")
        for start_date, end_date in data['shooting_dates']:
            print(f"  - Start Date: {start_date}, End Date: {end_date}")
        print("Locations:")
        for location in data['locations']:
            print(f"  - {location}")
        print()
        counter += 1

    print(f"Displayed information for 20 of {len(movie_data)} movies.")

# Handle exceptions
except FileNotFoundError:
    print(f"JSON file not found at {json_file_path}")

except json.JSONDecodeError as e:
    print(f"Error parsing JSON data: {e}")
```

2. Analyze the resulting CSV file to display, for all entries: title, author, and total number of loans.

```python
import csv

csv_file_path = "paris_most_loaned_titles.csv"  # Replace with the path to your CSV file

try:
    # Open and read the CSV file
    with open(csv_file_path, "r", encoding="utf-8", newline="") as file:
        csv_reader = csv.reader(file, delimiter=";")

        # Skip the header row containing the column titles
        next(csv_reader)

        # Create a dictionary to store loan data
        loan_data = {}

        # Process each row in the CSV file
        for row in csv_reader:
            title = row[3]
            author = row[4]
            total_loans = int(row[6]) if row[6].isdigit() else 0

            if title not in loan_data:
                loan_data[title] = {
                    "author": author,
                    "total_loans": total_loans,
                }
            else:
                loan_data[title]["total_loans"] += total_loans

        # Display information for all entries
        for title, data in loan_data.items():
            print(f"Title: {title}")
            print(f"Author: {data['author']}")
            print(f"Total Number of Loans: {data['total_loans']}")
            print()

    print(f"Displayed information for {len(loan_data)} entries.")

# Handle exceptions
except FileNotFoundError:
    print(f"CSV file not found at {csv_file_path}")

except csv.Error as e:
    print(f"Error reading the CSV file: {e}")
```

4. Display titles in order of profitability (in descending order of the number of loans per copy).

```python
import csv

csv_file_path = "paris_most_loaned_titles.csv"  # Replace with the path to your CSV file

try:
    # Open and read the CSV file
    with open(csv_file_path, "r", encoding="utf-8", newline="") as file:
        csv_reader = csv.reader(file, delimiter=";")

        # Skip the header row
        next(csv_reader)

        # Create a dictionary to store loan data for each title
        title_data = {}

        # Process each row in the CSV file
        for row in csv_reader:
            title = row[3]
            total_loans = int(row[6]) if row[6].isdigit() else 0
            total_copies = int(row[]) if row[8].isdigit() else 0

            # Calculate loans per copy
            loans_per_copy = total_loans / total_copies if total_copies > 0 else 0

            # Store loans per copy in the title data
            title_data[title] = {
                "total_loans": total_loans,
                "total_copies": total_copies,
                "loans_per_copy": loans_per_copy,
            }

        # Sort titles by profitability (loans per copy) in descending order
        sorted_titles = sorted(title_data, key=lambda x: title_data[x]["loans_per_copy"], reverse=True)

        # Display information for each title
        for title in sorted_titles:
            data = title_data[title]
            print(f"Title: {title}")
            print(f"Total Number of Loans: {data['total_loans']}")
            print(f"Total Number of Copies: {data['total_copies']}")
            print(f"Loans per Copy: {data['loans_per_copy']:.2f}")
            print()

    print(f"Displayed information for {len(title_data)} titles in order of profitability.")

# Handle exceptions
except FileNotFoundError:
    print(f"CSV file not found at {csv_file_path}")

except csv.Error as e:
    print(f"Error reading the CSV file: {e}")
```
